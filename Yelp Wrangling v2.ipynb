```python
### Tokenizing text into list of list

import nltk 
from nltk.tokenize import RegexpTokenizer

tokens =[]
for sentence in df_tip.text:
    #print (sentence)
    tokens.append(nltk.word_tokenize(sentence))
#print (tokens)

flattened_tokens = [y for x in tokens for y in x]   
print(flattened_tokens)
```

```python
### Remove punctuations, numbers and change to lowercase in the list 

import re

### flatten list of list into single list
flattened_tokens = [y for x in tokens for y in x]
#print(flattened_tokens)

def preprocess(text):
    clean_data = []
    for x in (text[:]): #this is Df_pd for Df_np (text[:])
        new_text = re.sub('<.*?>', '', x)   # remove HTML tags
        new_text = re.sub(r'[^\w\s]', '', new_text) # remove punc.
        new_text = re.sub(r'\d+','',new_text)# remove numbers
        new_text = new_text.lower() # lower case, .upper() for upper          
        if new_text != '':
            clean_data.append(new_text)
    return clean_data             


preprocess_list = preprocess(flattened_tokens)
print(preprocess_list)
```

```python
### Finding the number of positive and negative words in the reviews

import textblob
from textblob import TextBlob

def word_polarity(test_subset):
    pos_word_list=[]
    neu_word_list=[]
    neg_word_list=[]

    for word in test_subset:               
        testimonial = TextBlob(word)
        if testimonial.sentiment.polarity >= 0.5:
            pos_word_list.append(word)
        elif testimonial.sentiment.polarity <= -0.5:
            neg_word_list.append(word)
        else:
            neu_word_list.append(word)

    #print('Positive :',pos_word_list) 
    #print(len(pos_word_list))
    print(f"The number of positive words in the reviews are: {len(pos_word_list)}")
    #print('Neutral :',neu_word_list) 
    #print(len(neu_word_list))
    print(f"The number of neutral words in the reviews are: {len(neu_word_list)}")
    #print('Negative :',neg_word_list)
    #print(len(neg_word_list))
    print(f"The number of negative words in the reviews are: {len(neg_word_list)}")

word_polarity(preprocess_list)
```
